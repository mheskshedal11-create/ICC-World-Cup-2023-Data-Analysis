<p>
The first function, <b>OverallBatsmanStats</b>, calculates the overall tournament statistics for each batsman:
</p>

<ul>
    <li><b>Player:</b> Name of the player</li>
    <li><b>Matches:</b> Total number of matches played</li>
    <li><b>Runs:</b> Total number of runs scored</li>
    <li><b>Average:</b> Batting average</li>
    <li><b>Strike Rate:</b> Batting strike rate</li>
    <li><b>Balls Faced:</b> Total number of balls faced</li>
    <li><b>Highest Score:</b> Highest score achieved in a match</li>
    <li><b>50s:</b> Total number of half-centuries scored</li>
    <li><b>100s:</b> Total number of centuries scored</li>
    <li><b>Dismissals:</b> Number of times the player got out</li>
</ul>

```python
def OverallBatsmanStats(player_name):

    player = player_name

    playerList = df['striker'].unique().tolist()

    # Check whether player exists or not
    if player in playerList:

        batsman = df[df['striker'] == player]

        # Total matches played
        matches = batsman['match_id'].nunique()

        # Total runs scored
        score = batsman['runs_off_bat'].sum()

        # Runs scored in each match
        runs = batsman.groupby('match_id')['runs_off_bat'].sum().values.tolist()

        # Count 50s and 100s
        fifties = 0
        hundreds = 0

        for i in runs:
            if i >= 50 and i < 100:
                fifties += 1
            elif i >= 100:
                hundreds += 1

        # Highest score
        highestScore = batsman.groupby('match_id')['runs_off_bat'].sum().max()

        # Number of dismissals
        dismissed = batsman['player_dismissed'].count()

        # Balls faced
        ballsFaced = batsman['runs_off_bat'].count()

        # Strike Rate
        sr = (score / ballsFaced) * 100 if ballsFaced > 0 else 0

        # Batting Average
        avg = score / dismissed if dismissed > 0 else score

        # Final DataFrame
        data = {
            'PlayerName': [player],
            'Matches': [matches],
            'Runs': [score],
            'Average': [round(avg, 2)],
            'StrikeRate': [round(sr, 2)],
            'BallsFaced': [ballsFaced],
            'HighestScore': [highestScore],
            '50s': [fifties],
            '100s': [hundreds],
            'Dismissed': [dismissed]
        }

        Stats_df = pd.DataFrame(data)

        return Stats_df

    else:
        return 'Invalid Player Name. Please check the spelling and try again.'
```

">
The second function, <b>AgainstEachTeamBatsmanAnalysis</b>, calculates the batsman's performance against each team in a single match:
</p>

<ul>
    <li><b>Scores:</b> Total number of runs scored in the match</li>
    <li><b>Balls:</b> Total number of balls faced in the match</li>
    <li><b>Opponent:</b> The opponent team played against</li>
    <li><b>Venue:</b> The stadium or ground where the match was played</li>
</ul>

```python
def AgainstEachTeamBatsmanAnalysis(p_name):
    player = p_name
    
    playerList = df['striker'].unique().tolist()
    
    #Checking if the player name is in list of players who played ICC CWC_2023
    if player in playerList:
        
        batsman = df[df['striker'] == player]
        
        matches_played = sorted(batsman['match_id'].unique().tolist())
        
        matches_played
                
        scores = []
        balls = []
        opponent = []
        venue = []

        for i in matches_played:
            scores.append(batsman[batsman['match_id'] == i]['runs_off_bat'].sum())
            balls.append(batsman[batsman['match_id'] == i]['runs_off_bat'].count())
            opponent.append(batsman[batsman['match_id'] == i]['bowling_team'].unique()[0])
            venue.append(batsman[batsman['match_id'] == i]['venue'].unique()[0].split(',')[-1])
        
        
        teamWise = {
            'Scores' : scores,
            'Balls' : balls,
            'Opponent' : opponent,
            'Venue' : venue
        }
        
        
        TeamWiseStats = pd.DataFrame(teamWise)
        
        return TeamWiseStats
    
    else:
        return 'Invalid Player Name, Please Re-Enter the name of Player or Check the spelling of the name'
```

<p>
The first function, <b>OverallBowlerStats</b>, calculates the overall tournament statistics for each bowler:
</p>

<ul>
    <li><b>Player:</b> Name of the bowler</li>
    <li><b>Matches:</b> Total number of matches played</li>
    <li><b>Overs/Balls:</b> Total number of balls or overs bowled</li>
    <li><b>Runs Conceded:</b> Total runs given by the bowler</li>
    <li><b>Wickets:</b> Total number of wickets taken</li>
    <li><b>Economy Rate:</b> Runs conceded per over</li>
    <li><b>Dot Balls:</b> Number of balls where no runs were scored</li>
    <li><b>4s Conceded:</b> Number of boundaries (4 runs) given</li>
    <li><b>6s Conceded:</b> Number of sixes given</li>
    <li><b>No Balls:</b> Total number of no-ball deliveries bowled</li>
    <li><b>Wides:</b> Total number of wide deliveries bowled</li>
</ul>

```python
def OverallBowlerStats(player):
    player=player

    ListOfBowler=df['bowler'].unique().tolist()

    if player in ListOfBowler:

        Matches = df[df['bowler'] == player].groupby('match_id')['bowler'].count()
        run=df[df['bowler']==player].groupby('bowling_team')[['runs_off_bat','extras']].sum()
        total=pd.DataFrame(run)
        total_runs=(total['runs_off_bat'] + total['extras']).sum()
        wicket=df[df['bowler']==player]['wicket_type'].count()

        #for econamy
        total_ball=df[df['bowler']==player]['ball'].count()
        over=total_ball/6
        economy=round(total_runs/over,2)

        dot_ball=df[(df['bowler']==player)&(df['runs_off_bat']==0)&(df['extras']==0)]['ball'].count()
        s4=df[(df['bowler']==player)&(df['runs_off_bat']==4)]['runs_off_bat'].count()
        s6=df[(df['bowler']==player)&(df['runs_off_bat']==6)]['runs_off_bat'].count()
        no_ball=df[df['bowler']==player]['noballs'].sum()
        wides=df[df['bowler']==player]['wides'].sum()
    
    
        overview = {
            'Player_name': player,
            'total_match': len(Matches),
            'Total Run':total_runs,
            'Wicket':wicket,
            'Economy':economy,
            'Dot Balls':dot_ball,
            '4s Conceded':s4,
            '6s Conceded':s6,
            'No Balls':no_ball,
            'Wides':wides,
            'Total Over':over

        }
    
        fr = pd.DataFrame([overview])
        return fr
    else:
        return 'Invalid Player Name. Please check the spelling and try again.'
```
